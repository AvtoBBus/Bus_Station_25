### Dyhotomy function

```python
def dyhotomy(a, b, delta):
    dyhotomyCallCount = 0
    result = (a + b) / 2

    eps = delta / 100

    while abs(b - a) > 2 * delta:
        
        mid = (a + b) / 2
        dyhotomyCallCount += 2

        result = mid
        if mainFunction(mid - eps) <= mainFunction(mid + eps):
            b = mid + eps
        else:
            a = mid - eps

    if result is None:
        print('Something went wrong')
    else:
        print(f'Result = {(a + b) / 2}\nError = {(b - a) / 2}\nFunction call = {dyhotomyCallCount}')
```

```python
dyhotomy(A, B, DELTA)
```

### Fibonacci

```python
def fibonacci(x):
    if x in {0, 1}:
        return x
    return fibonacci(x - 1) + fibonacci(x - 2)
```
